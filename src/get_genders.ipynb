The goal is to classify each song's author by male, female, band or collaboration. We will do so by using an API.

```python
import pandas as pd

# Carreguem els datasets
high_pop = pd.read_csv('../Practica 1/high_popularity_spotify_data.csv')
low_pop = pd.read_csv('../Practica 1/low_popularity_spotify_data.csv')

# Afegim la columna de font per a l'anàlisi posterior
high_pop['popularity_type'] = 'High'
low_pop['popularity_type'] = 'Low'

# Fusionem (tenint en compte les ~5.000 cançons que esmentes a la pràctica)
df_total = pd.concat([high_pop, low_pop], ignore_index=True) # [cite: 12]

df_total
```

```python
# Artistes únics

import numpy as np

# 4. Crear llista d'artistes únics per a l'enriquiment
unique_artists = sorted(list(set(df_total['track_artist'].dropna().unique())))
unique_artists_df = pd.DataFrame({'artist_name': unique_artists, 'artist_class': np.nan})

# 5. Lògica d'eficiència: Marcar col·laboracions directament
# Si el nom conté una coma, marquem com a 'collaboration' i no consultarem l'API
unique_artists_df.loc[unique_artists_df['artist_name'].str.contains(','), 'artist_class'] = 'collaboration'

# Guardar el fitxer de treball
unique_artists_df.to_csv('unique_artists.csv', index=False)
print(f"Fitxer creat amb {len(unique_artists_df)} artistes.")
print(f"Col·laboracions detectades (no requeriran API): {unique_artists_df['artist_class'].count()}")
```

```python
import musicbrainzngs
import pandas as pd
import time

# Configuració de l'API
musicbrainzngs.set_useragent("ProjecteMusicaTeo", "1.1", "el-teu-email@estudiant.com")

def classify_artist_gender_or_type(artist_name, current_classification):
    # 1. Identificació de col·laboracions per text
    if current_classification == "collaboration":
        return "collaboration"
    
    try:
        # 2. CERCA de l'artista per obtenir el seu ID (MBID)
        print(f"Cercant ID per a: {artist_name}...")
        search_result = musicbrainzngs.search_artists(artist=artist_name, limit=1)
        
        if not search_result['artist-list']:
            return "unknown"
        
        # Agafem el primer resultat i el seu ID
        artist_id = search_result['artist-list'][0]['id']
        
        # 3. CONSULTA DIRECTA per ID (això ens dona la fitxa completa)
        # Afegim una pausa per no saturar l'API
        time.sleep(1) 
        full_info = musicbrainzngs.get_artist_by_id(artist_id)['artist']
        
        # Extraiem el tipus i el gènere amb normalització a minúscules
        artist_type = full_info.get('type', '').lower()
        gender = full_info.get('gender', 'unknown').lower()
        
        # CLASSE: Band
        if artist_type in ['group', 'orchestra', 'choir']:
            return "band"
        
        # CLASSE: Male / Female
        if gender in ['male', 'female']:
            return gender
        
        return "unknown"

    except Exception as e:
        return f"error: {str(e)}"
    finally:
        time.sleep(0.5) # Pausa addicional de seguretat

# Execució del test amb les teves 5 files
df = pd.read_csv('unique_artists.csv')
df_test = df.head(10).copy()

print("Iniciant procés de classificació profunda...")
df['artist_class'] = df.apply(lambda row: classify_artist_gender_or_type(row['artist_name'], row['artist_class']), axis=1)

print("\n--- RESULTAT DEL TEST CORREGIT ---")
print(df[['artist_name', 'artist_class']])

df.to_csv('unique_artists_classified.csv', index=False)
```

```python
import pandas as pd

# Carreguem el diccionari d'artistes ja enriquits amb l'API i les col·laboracions
df_classes = pd.read_csv('unique_artists_classified.csv')

# Realitzem la fusió (Merge)
# Unim el dataset de cançons amb el de classes usant el nom de l'artista com a clau
df_final = pd.merge(
    df_total, 
    df_classes, 
    left_on='track_artist',   # Columna al dataset de Spotify
    right_on='artist_name',   # Columna al teu diccionari d'enriquiment
    how='left'                
)

# Eliminar la columna repetida 'artist_name'
if 'artist_name' in df_final.columns:
    df_final = df_final.drop(columns=['artist_name'])

# Guardar el dataset definitiu per a la visualització
df_final.to_csv('dataset_musical_final_per_viz.csv', index=False)

print(f"Fusió completada!")
print(f"Total de cançons: {len(df_final)}")
print("Resum de classificacions:")
print(df_final['artist_class'].value_counts())
```
